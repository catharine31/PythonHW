### map

```python
items = [500, 30, 2000]
c = 30.8594
exchange = []
for i in items:
    e = i * c
    exchange.append(e)
print(exchange)
```

```python
def exchange(p):
    return p * c
exchange(500)
    
```

```python
#map(函數, List)
map(exchange, items)
```

```python
list(map(exchange, items))
```

### [Lambda] 給他x乘上c作用在items這個list上

```python
list(map(lambda x: x * c, items))
```

### [List comprehesion] 把items裡的東西一個一個拿出來p*c

```python
[p * c for p in items]
```

### string格式

```python
name = "Becca"
age = 20
print("格式：format()")
print("您好, 我是 {}, 今年 {} 歲".format(name, age))
print("格式：%")
print("您好, 我是%s 今年 %d 歲"%(name, age))
print("f-string")
print(f"您好，我是{name}, 今年{age}歲")
```

```python
c = 30.8594353
print(f"1美金和台幣{c}元。")
print(f"1美金和台幣{c:.2f}元。")
print(f"1美金和台幣{c:10.2f}元。")
```

```python
strs = f"> <"
spam = f"！{strs}！"
print(spam)
spam1 = f"！{strs:20s}！"
print("靠左", spam1)
spam2 = f"！{strs:>20s}！"
print("靠右", spam2)
spam3= f"！{strs:^20s}！"
print("靠右", spam3)
```

```python
for i in range(1, 6):
    s = "*"*(2*i - 1)
    print(f"{s:^21s}")
```
