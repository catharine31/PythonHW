## 插播：contour用法

```python
Z = np.random.randint(1, 3, (4, 4))
```

```python
Z
```

```python
xL = np.array([1, 2, 3, 4])
yL = np.array([5, 6, 7, 8])
X, Y = np.meshgrid(xL, yL)
```

```python
plt.contour(X, Y, Z)
plt.scatter(X.ravel(), Y.ravel(), c=Z.ravel())
```

## 鳶尾花數據庫

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
# print(iris.DESCR)
```

```python
x = iris.data
y = iris.target
```

```python
x[:5]
```

```python
y[:5]
```

```python
X = x[:, :2]
```

```python
X
```

```python
Y = y
```

```python
len(X)
```

## 區分訓練資料、預測資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,  y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap="Paired")
```

## STEP1 開啟數據學習機

```python
clf = SVC(gamma="auto")
```

## STEP2 訓練

```python
clf.fit(x_train, y_train)
```

## STEP 3預測

```python
y_predict = clf.predict(x_test)
```

```python
y_test
```

```python
y_predict - y_test
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_predict - y_test)
```

```python
x0 = np.arange(3.8, 8.2, 0.02)
y0 = np.arange(1.8, 4.7, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, cmap="Paired", alpha=0.6)
plt.scatter(x_test[:, 0], x_test[:, 1], cmap="Paired", c=y_test)
```
