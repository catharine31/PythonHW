挑戰題一：買相機

```python
months = ["1", "2", "3", "4"]
price = [100, 150, 230, 80]
plt.xlabel("Months")
plt.ylabel("k")
total = 0
priceTotal = []
for i in price:
    total += i
    priceTotal.append(total)
plt.plot(months, priceTotal)
```

```python
plt.pie(price, labels = months, autopct = "%1.1f%%")
```
