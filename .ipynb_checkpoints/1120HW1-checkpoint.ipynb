```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Z = np.random.randint(1, 3, (4, 4))
```

```python
A = np.random.randint(0, 6, 3)

x = np.linspace(0, A[0]*np.pi)
y = np.linspace(A[1]**np.pi, A[2]**np.pi)
[X,Y] = np.meshgrid(x,y)
Z = np.cos(X)+np.sin(Y)
plt.contour(X,Y,Z)
```

```python
plt.contourf(X, Y, Z)
plt.scatter(X.ravel(), Y.ravel(), c=Z.ravel())
```
