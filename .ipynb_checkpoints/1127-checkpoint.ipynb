```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## PCA 通常用在壓縮維度

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
from sklearn.decomposition import PCA
```

## step 1. 開PCA機

```python
# 壓到二維
pca = PCA(n_components=2)
```

## step 2. 訓練

```python
pca.fit(x)
```

## step 3. 預測

```python
x[87]
```

```python
X = pca.transform(x)
```

```python
X[87]
```

```python
# c=y，顏色依分類來分，分類就在y裡面
plt.scatter(X[:, 0], X[:, 1], c=y)
```

## 準備訓練/測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## step1: 開 SVC 函數學習機

```python
# SVC分類
from sklearn.svm import SVC
clf = SVC(gamma='auto')
```

## step2: 訓練

```python
clf.fit(x_train, y_train)
```

## step3: 預測

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
x0 = np.arange(-4, 4.2, 0.02)
y0 = np.arange(-1.5, 1.7, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3, cmap='Paired')
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap='Paired')
```

## K-Means自動分類

```python
x = np.random.rand(100, 2)
plt.scatter(x[:, 0], x[:, 1])
```

## step1: 打開 K-Means學習機

```python
from sklearn.cluster import KMeans
```

```python
# 分3類
clf = KMeans(n_clusters=3)
```

## step2: 學習

```python
clf.fit(x)
```

## step3: 預測

```python
clf.predict(x)
```

```python
clf.labels_
```

```python
# 分三類
plt.scatter(x[:, 0], x[:, 1], c=clf.labels_)
```

```python
x0 = y0 = np.arange(-0.2, 1.2, 0.02)
xm, ym = np.meshgrid(x0, y0)

P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3, cmap='Paired')
plt.scatter(x[:, 0], x[:, 1], c=clf.labels_, cmap='Paired')
```

## make_bpobs生更像真實世界的數據

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
# 取500筆資料，分三類，2個輸入的變數
x, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=0)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='Paired')
```

```python
plt.scatter(x[:, 0], x[:, 1])
```

## step1: 打開 K-Means學習機

```python
clf = KMeans(n_clusters=3)
```

## step2: 學習

```python
clf.fit(x)
```

## step3: 預測

```python
plt.scatter(x[:, 0], x[:, 1], c=clf.labels_, cmap='Paired')
```

## Mean-Shifts完全自動分類
不知道要分幾類，叫電腦自己分類

## step1: 打開 K-Means學習機

```python
from sklearn.cluster import MeanShift
```

```python
clf = MeanShift()
```

## step2: 學習

```python
clf.fit(x)
```

## step3: 預測

```python
plt.scatter(x[:, 0], x[:, 1], c=clf.labels_, cmap='Paired')
```

```python
x0 = np.arange(-4.5, 5.2, 0.02)
y0 = np.arange(-3, 7, 0.02)
xm, ym = np.meshgrid(x0, y0)

P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3, cmap='Paired')
plt.scatter(x[:, 0], x[:, 1], c=clf.labels_, cmap='Paired')
```

## cross_val_score

```python
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='Paired')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto')
```

```python
# 訓練資料和操是資料分5次
scores = cross_val_score(clf, x, y, cv=5)
```

```python
scores.max()
```

```python
scores.mean()
```
